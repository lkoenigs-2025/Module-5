```python
def load_words(filename: str) -> list[str]:
  """
  Reads a file containing one word per line.
  Returns a list of words.
  """
  my_list = []

  with open(filename,"r") as readfile:
    line = readfile.readline()
    line = line.strip()
    while line != "":
      my_list.append(line)
      line = readfile.readline()
      line = line.strip()

  return my_list


def normalize(text: str) -> str:
  """
  Returns lowercase letters only (a-z).
  Removes punctuation, spaces, digits, etc.
  """
  new = ""
  for i in text:
    if i.isalpha():
      new += i.lower()
    else:
      continue
  return new


def set_signature(text: str) -> frozenset:
  """
  Normalize the text, then return frozenset of unique letters.
  Example: "aab" -> frozenset({'a','b'})
  """
  clean = normalize(text)
  letters = set(clean)
  word = frozenset(letters)
  return word



def freq_signature(text: str) -> tuple:
  """
  Normalize the text, then return a 26-length tuple of counts for a..z.
  Example: "aab" -> (2,1,0,0,...)
  """

  clean = normalize(text)
  lst_letters = [0] *26
  alpha = "abcdefghijklmnopqrstuvwxyz"
  for i in clean:
    for j in range(len(alpha)):
      if i == alpha[j]:
        lst_letters[j] +=1
        break
  tpl = tuple(lst_letters)
  return tpl


def is_anagram_using_sets(a: str, b: str) -> bool:
  return set_signature(a) == set_signature(b)


def is_anagram(a: str, b: str) -> bool:
  return freq_signature(a) == freq_signature(b)


def group_anagrams(words: list[str]) -> dict[tuple, list[str]]:
  """
  Groups words by freq_signature (correct).
  key = freq_signature
  value = list of original words with that signature
  """
  my_dict = {}
  num = 0
  key_count = 0
  singletons = 0
  largest = 0
  all_values = []
  top_five = []
  longest = 0

  for i in words:
    tpl = freq_signature(i)
    if tpl not in my_dict:
      my_dict[tpl] = [i]
    else:
      my_dict[tpl].append(i)
    num+=1

  for key,value in my_dict.items():
    key_count+=1
    if len(value) == 1:
      singletons +=1
    if len(value) > largest:
      largest = len(value)
      lrg_group = value
    all_values.append(value)

  for five in range(5):
    for i in all_values:
      if len(i) > longest:
        longest = len(i)
        longest_lst = i
    top_five.append(longest_lst)
    all_values.remove(longest_lst)
    longest = 0


  print("ANAGRAM GROUP REPORT")
  print(f"Total words loaded: {num}")
  print(f"Total anagram groups: {key_count}")
  print(f"Singleton groups (size 1): {singletons}")

  print(f"Largest group (size {largest}): {lrg_group}")


  print(f"Top 5 group sizes:")
  for i in top_five:
    size = len(i)
    print(f"Size {size}: {i}")

  return my_dict



def main():
  my_list = load_words("words.txt")

  print("\nSET COUNTEREXAMPLES: (set method says True, but NOT anagrams)")
  print(f'adder vs read : {is_anagram_using_sets("adder", "read")}')
  print(f'mississippi vs misp : {is_anagram_using_sets("mississippi", "misp")}')
  print(f'aab vs ab : {is_anagram_using_sets("aab", "ab")}')

  print("\nCORRECT EXAMPLE:")
  print(f'cuepack vs cupcake : {is_anagram("cuepack", "cupcake")}')


  group_anagrams(my_list)
  print("\nAnswer for 1. X is O(1) because in a set it just goes to where x is.")
  print("Answer for 2. X is O(n) in a list because it goes through the items one by one until it finds x.")
  print("Answer for 3. A freq signature for a word of length k is O(k) because it searches through the whole word.")
  print("Answer for 4. Sets fail for anagrams because it takes away any duplicates the words might have, making the final answer incorrect.")



main()
```
