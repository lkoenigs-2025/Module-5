```python
class Pet:

  country_of_origin = "USA"
  num_pets = 0

  def __init__(self, name: str, age: int, owner: str)-> None:
    self.__name = name
    self.__age = age
    self.__owner = owner
    Pet.count_pets()

  def get_name(self)->str:
    return self.__name

  def get_age(self)->str:
    return self.__age

  def get_owner(self)-> str:
    return self.__owner

  def set_name(self, name:str)-> None:
    self.__name = name

  def set_age(self, age:int)-> None:
    self.__age = age

  def set_owner(self, owner:str)-> None:
    self.__owner = owner

  @classmethod
  def count_pets(cls):
    cls.num_pets +=1

  #__str__: return f string and it will give u a string,
  #and when u print any objet it will call it automatically

  def __str__(self):
    return f"{self.__name} {self.__age} {self.__owner}"

#Create two pets
p1 = Pet("Fifi", "2", "Reizi")
p2 = Pet("Rufus", "5", "Ruchama")


print(p1.get_age())
print(p2.get_age())
print(p1.get_name())
print(p2.get_name())

p2.country_of_origin = "France"

print(p1.country_of_origin)
print(p2.country_of_origin)

print(p1.num_pets)

#will print str instead of the type
print(p1)
```
