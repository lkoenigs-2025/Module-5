```python
store = {}
store

def add_product(store, name, price, stock):
  if not isinstance(store, dict):
    return False
  if name in store:
    return False
  store[name] = {"price": price, "stock": stock, "reviews": {}}
  return True

  # Tests
print(add_product(store, "Laptop", 1200, 5))   # expected True
print(add_product(store, "Laptop", 1200, 5))   # expected False
print(store)
```

```python
def add_review(store, name, user, score):
  if not isinstance(store, dict):
    return False
  if name not in store:
    return False
  if "reviews" not in store[name] or not isinstance(store[name]["reviews"], dict):
    store[name]["reviews"] = {}
  if score > 10 or score < 1:
    return False
  store[name]["reviews"][user] = score
  return True

# Tests
print(add_review(store, "Laptop", "Alice", 9))   # expected True
print(add_review(store, "Laptop", "Bob", 8))     # expected True
print(add_review(store, "Phone", "Tom", 7))      # expected False
print(add_review(store, "Laptop", "Eve", 15))    # expected False
print(store)
```

```python
def get_average_score(store, name):
  if not isinstance(store, dict):
    return None
  if name not in store or not isinstance(name,str) or not isinstance(store[name], dict):
    return None
  if store[name]["reviews"] == {}:
    return None
  total = 0
  for value in store[name]["reviews"].values():
    total += value
  average = total / len(store[name]["reviews"])
  return average
# Tests
print(get_average_score(store, "Laptop"))   # expected 8.5
print(get_average_score(store, "Phone"))    # expected None
```

```python
def group_products_by_price_range(store):
  if not isinstance(store, dict):
    return {}

  prices = {}
  prices["cheap"] = []
  prices["expensive"] = []
  prices["mid"] = []
  for name in store:
    if "price" in store[name] or isinstance(store[name]["price"], float):
      if store[name]["price"] < 50:
          prices["cheap"].append(name)

      elif store[name]["price"] > 500:
          prices["expensive"].append(name)

      else:
        prices["mid"].append(name)
  return prices
# Setup
add_product(store, "Mouse", 25, 50)
add_product(store, "Monitor", 300, 10)

# Test
print(group_products_by_price_range(store))
```

```python
def validate_store(store):
  problems = []
  if not isinstance(store, dict):
    problems.append("Store not a dict")
  for name in store:
    if not isinstance(name,str):
      problems.append("product name is not a string")
    if not isinstance(store[name], dict):
      problems.append("product is not a dict")
    if "price" not in store[name] or "stock" not in store[name] or "reviews" not in store[name]:
      problems.append("product missing at least one key")
    if not isinstance(store[name]["price"],(int,float)):
      problems.append("price is not a float or an int")
    if not isinstance(store[name]["stock"],int):
      problems.append("stock is not an integer")
    if not isinstance(store[name]["reviews"],dict):
      problems.append("reviews not a dict")
    for user in store[name]["reviews"]:
      if not isinstance(user,str):
        problems.append("user is not a string")
      if store[name]["reviews"][user] > 10 or store[name]["reviews"][user] < 1:
        problems.append("score is not in between 1 and 10")
  return problems
  # Tests
print("Problems:", validate_store(store))
```

```python
def update_stock(store, name, amount):
  if not isinstance(store, dict):
    return False
  if name not in store:
    return False
  if not isinstance(amount, int):
    return False
  store[name]["stock"] += amount
  return True

  # Tests
print(update_stock(store, "Laptop", 3))    # expected True
print(update_stock(store, "Laptop", -2))   # expected True
print(update_stock(store, "Phone", 5))     # expected False
print(store)
```

```python
def find_low_stock(store, threshold):
  products = []
  if not isinstance(store, dict):
    return []
  for name in store:
    if not isinstance(name, str) and not isinstance(store[name],dict) and "stock" not in store[name]:
      if store[name]["stock"] < threshold:
        products.append(name)
  return products
# Tests
print(find_low_stock(store, 10))   # expected ["Laptop"] (depending on updates)
```

```python
def unique_count(nums):
  if not isinstance(nums, list):
    return 0
  my_set= len(set(nums))
  return my_set
# Tests (do not edit)
print(unique_count([1,2,2,3,3,3]))   # expected 3
print(unique_count([]))              # expected 0
print(unique_count("abc"))           # expected 0
```

```python
def has_common_element(a, b):
  if not isinstance(a, list) or not isinstance(b, list):
    return False
  set_a = set(a)
  set_b = set(b)
  if set_a.intersection(set_b):
    return True
  return False
# Tests
print(has_common_element([1,2,3], [3,4,5]))   # expected True
print(has_common_element([1,2], [3,4]))       # expected False
print(has_common_element([], [1]))            # expected False
```

```python
def remove_duplicates_preserve_order(nums):
  if not isinstance(nums,list):
    return []
  final = set()
  new = []
  for i in nums:
    if i not in final:
      new.append(i)
      final.add(i)
  return new

# Tests
print(remove_duplicates_preserve_order([1,2,2,3,1]))  # expected [1,2,3]
print(remove_duplicates_preserve_order([]))           # expected []
```

```python
def first_repeating(nums):
  if not isinstance(nums,list):
    return None
  seen = set()
  for i in nums:
    if i not in seen:
      seen.add(i)
    else:
      return i
  return None
# Tests
print(first_repeating([2,1,3,5,3,2]))   # expected 3
print(first_repeating([1,2,3]))         # expected None
```

```python
def missing_number(nums):
  if not isinstance(nums,list):
    return None
  sort = sorted(nums)
  for i in sort:
    if i+1 not in sort:
      return i+1
# Tests
print(missing_number([3,0,1]))   # expected 2
print(missing_number([0,1]))     # expected 2
```

```python
def are_disjoint(a, b):
  if not isinstance(a,list) or not isinstance(b, list):
    return False
  sa = set(a)
  sb = set(b)
  if not sa.intersection(sb):
    return True
  return False

# Tests
print(are_disjoint([1,2,3], [4,5]))   # expected True
print(are_disjoint([1,2,3], [3,4]))   # expected False
```

```python
def count_distinct_pairs(nums, target):
  my_set = set(nums)
  new = []
  for i in my_set:
    value = target - i
    if value in my_set:
      if (i, value) not in new and (value,i) not in new:
        new.append((i, value))
  return len(new)


# Tests
print(count_distinct_pairs([1,2,3,4,3], 6))   # expected 2  → (2,4), (3,3)
print(count_distinct_pairs([1,1,1], 2))       # expected 1
```

```python
def longest_streak_start(nums):
  if len(nums) == 0:
    return None
  seen = set()
  longest = 0
  for i in nums:
    string = str(i)
    if i not in seen:
      seen.add(i)
      if len(string) > longest:
        longest = (len(string))
        return len(string[0])

# Tests
print(longest_streak_start([100,4,200,1,3,2]))   # expected 1
print(longest_streak_start([]))                  # expected None
```
