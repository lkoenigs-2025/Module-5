````python
school = {
   "Alice": {
       "age": 20,
       "major": "CS",
       "grades": {
           "Math": 90,
           "CS": 95
       }
   },
   "Bob": {
       "age": 19,
       "major": "History",
       "grades": {
           "History": 88
       }
   },
   "Dina": {
       "age": 21,
       "major": "Math",
       "grades": {
           "Math": 100,
           "CS": 84
       }
   }
}
````

````python
def add_student(school, student, age, major):
  if not isinstance(school,dict):
    raise TypeError("school must be a dict")
  if not isinstance(student, str) or student.strip() == "":
    raise ValueError("student must be a string and not empty")
  if not isinstance(age,int) or not  0 < age <= 25:
    raise ValueError("age is not correct range")
  if not isinstance(major, str) or major.strip() == "":
    raise ValueError("Major must be a string and not empty")

  if student not in school:
    school[student] = {"age": age, "major": major, "grades": {}}
    return school
  if not isinstance(school[student], dict):
    raise TypeError(f'{student} value should be a dict but is corrupted')
  if "age" not in school[student]:
    school[student]["age"] = age
  if "major" not in school[student]:
    school[student]["major"] = major
  if "grades" not in school[student]:
    school[student]["grades"] = {}

  if not isinstance(school[student]["grades"], dict):
    raise TypeError(f'Grades for {student} must be a dict')

  print(f'we just entered {school[student]}')
add_student(school, "Eliana", 19, "History")
````

## Building our CSV file named "student_data.csv"

````python
data = ["name", "age", "major", "course", "grade",
        "Alice", 20, "CS", "Math", 90,
        "Alice", 20, "CS", "CS", 88,
        "Bob", 21, "Math", "Math", 85,
        "Bob", 21, "Math", "CS", 90]

count = 1
with open("student_data.csv", "w") as fout:
    for item in data:
        item = str(item)
        if count == len(data) :
            fout.write(item)
        elif count % 5 == 0:
            fout.write(item + "\n")
        else:
            fout.write(item + ",")   # cleaner CSV
        count += 1

student_list = {}

with open("student_data.csv", "r") as infile:
  line = infile.readline()
  header = line.strip()
  keys = header.split(",")
  for i in line:
    line = infile.readline()
    line = line.strip()
    values = line.split(",")
    if len(values) == 5 and len(keys) == 5:
      student_list[values[0]] = {keys[1]: values[1], keys[2]: values[2], "grades": {}}
      student_list[values[0]]["grades"][values[3]] = values[4]


  print(student_list)
````

````python
def load_csv_data():
   with open("student_data.csv", "r") as fin:
       header = [h.strip() for h in fin.readline().strip().split(",")]


       for line in fin:
           parts = [p.strip() for p in line.strip().split(",")] # strip the line
           if len(parts) != len(header): # expecting 5 elements in the line, if not reject and go on to next line
               continue


           name, age, major, course, grade = parts # unpack
           age = int(age)
           grade = int(grade)


           if name not in student_list:
               student_list[name] = {"age": age, "major": major, "grades": {}}


           student_list[name]["grades"][course] = grade


load_csv_data()
print(student_list)

````

## Playing with Sets()

````python

#return letters that are unique

#for letters that are seen exactly once
def exactly_once(items):
  answer = []
  unseen = set(items) #if any duplicates they will not be in the set

  for item in items:
    if item in unseen:
      unseen.remove(item)
      answer.append(item)

    else:
      if item in answer:
        answer.remove(item)
  return answer

print(exactly_once(["a","b","c","c","c","c","c","d","d"]))
````

## Leetcode 217 :
Given an integer array nums, return true if any value appears at least twice in the array, and return false if every element is distinct.

````python
# answer
def use_sets(nums):
  return len(set(nums)) != len(nums)
print(use_sets([1,5,4,2,4,2,4,2]))
print(use_sets([1,2,3,4,5,6,7,8]))
````
